# Tarea2: Control de Versiones con Git



Los sistemas de control de versiones, son herramientas que ayudan a la administración y manejo del historial de cambios que se realicen en colecciones de archivos y carpetas, por lo que tambien ayudan a que la colaboración sea más fácil de realizar.

Los sistemas de control de versiones hacen uso de metadata en la cual se almacena información como, quien hizo el cambio, cual fue el cambio realizado o mensajes que se pueden agregar para identificar los cambios que se realicen.

Cuando es utilizado por una sola persona es muy útil para para revisión de donde fueron realizados los cambios o para trabajar en desarrollos paralelos, y cuando es utilizado por por más de una persona es muy útil para saber quien hizo un cambio, en que línea se realizó el cambio y también para resolver conflictos que estos cambios puedan ocasionar. 



### Modelo de datos de Git

La interfaz de Git utiliza la línea de comandos para interactuar, tiene también algunas versiones gráficas pero usualmente los usuarios de esta herramienta concuerdan en que prefieren el uso de la línea de comandos, por lo que, para entender un poco mas como funciona Git, primero se explicará como trabaja Git, de esta forma se hará mas fácil después el uso de los comandos que tiene disponibles.


Git modela el historial de una colección de archivos y carpetas de un directorio que se especifique como un conjunto de snapshots o capturas. 

Git utiliza objetos para poder organizar la estructura, en el nivel más alto que puede ser un directorio, le da el nombre de "tree" y debajo de el pueden haber otros directorios o carpetas a las cuales también puede llamar "tree" o archivos a los que llama "blob". Una snapshot es el nivel más alto desde donde se inicia el seguimiento.

Para modelar el historial, Git utiliza una grafo aciclico de snapshots, es decir, que cada snashot hace referencia a un conjunto de padres que la preceden, y se habla de conjunto porque puede ser precedido por uno o por mas padres en el caso de que se combinen cambios de dos "branches" o ramas de desarrollo . A estas snapshots Git las llama commit.


Cada objeto en Git tiene una referencia para poder identificarlo la cual es llamada Hash, esta es una cadena de 40 caracteres hexadecimales. Estas referencias son punteros hacia los commits. Estas referencias pueden ser renombradas para que su acceso sea más fácil.

Git utiliza la referencia master, para apuntar al ultimo commit realizado en la rama principal. También utiliza una referencia especial que indica en donde se encuentra el puntero actualmente, esta referencia se llama HEAD. Se pueden crear o identificar otras referencias para poder moverse a travez de las ramas del historial, y se puede utilizar HEAD para identificar en la rama en la que se esta trabajando en cada momento.

Un repositorio Git son los objetos de datos y refencias, que son lo que conforman el modelo de datos de Git.

Git tiene un área muy util llamada staging area o "area de preparación", en esta área se puede indicar que modificaciones se quiere incluir en el snapshot antes de realizarlo, algo que es muy útil cuando se estan preparando actualizaciones de cambios ya sea en archivos o directorios.



### Interfaz de linea de comandos de Git

Dentro de los comandos mas importantes para hacer uso de Git tenemos:

**git help < comando >**: Este es el comando básico para visualizar la ayuda de Git, si no agregamos la opción de < commando > nos muestra las opciones que tenemos disponibles

```python
!git help
```

```python
##si colocamos alguna de las opciones git va internamente a buscar el archivo 
#donde guarda la información y en este caso lo muestra en el navegador por ser un archivo html
!git help branch  
```

**git init**:  Para crear un repositorio nuevo con datos almacenados en directorio .git


```python
!git init #inicializa un repositorio Git vacio
```

**git status**: Este comando indica el estatus actual, por ejemplo, de los commits pendientes, o de los archivos o directorios que esten pendientes de agregar al repositorio.

Ejemplo: en el repositorio de prueba muestra que hay un archivo que no se ha agregado al repositorio:




```python
!git status
```

**git add < nombre de algun archivo >**:  agrega un archivo al area de staging, que agregará al repositorio al momento que se haga el proximo commit. 

$ git add README.md

Cuando se agrega un archivo ahora aparece listo en el area de staging 


**git commit**: crea un nuevo commit y se guardan todos los cambios que se tengan en el area de staging 

En este caso se puede agregar un argumento mensaje para indicar que cambios se estan agregando en el ejemplo se colocó que se agregó el archivo README2.md y git lo agrega al repositorio.




**git log**

git log, sin colocar argumentos muestra historial del log en un texto plano.

git log --all --graph --decorate, si colocamos estos argumentos git nos muestra el historial del log de una manera visual como un grafo aciclico.




**git branch**: muestra las ramas o "branches" que se tienen en el repositorio 

En este ejemplo solo tenemos la rama principal



**git branch < nombre >**: crea una rama con el nombre que se indique

En este ejemplo se crea una rama con el nombre Prueba, si se ejecuta el comando para ver las ramas ahora ya aparece Prueba.


    
    


**git checkout < branch >**: con este comando nos posicionamos en la rama con la que se quiera trabajar, en el ejemplo anterior, se creó una rama pero aun queda posicionado el puntero en la rama principal master.

    

Si se ejecuta el comando git checkout Prueba, el puntero HEAD ahora se mueve a la nueva rama





Si se ejecuta el comando checkout con el argumento -b se crea la rama que se indique y de una vez posiciona el puntero HEAD sobre la rama creada, esto es util para no tener que ejecutar dos comandos, uno para crear la rama y otro para posicionarse en la nueva rama, se puede simplificar en un solo paso.
